```python
import numpy as np 
import pandas as pd 
import re # this is the regular expression library which helps us manipulate text (strings) fairly easily and intuitively
import nltk # this is the Natural Language Tool Kit which contains a lot of functionalities for text analytics
import matplotlib.pyplot as plt
import string # this is used for string manipulations
import matplotlib
```

```python
nltk.download('inaugural')
from nltk.corpus import inaugural
inaugural.fileids()
```

#### Roosevelt Speech

```python
inaugural.raw('1941-Roosevelt.txt')
```

###### Number of words - Roosevelt Speech

```python
num_words = nltk.corpus.inaugural.words('1941-Roosevelt.txt')
print('Number of words in Roosevelt speech are', len(num_words))
```

```python
num_chars = nltk.corpus.inaugural.raw('1941-Roosevelt.txt')
print('Number of characters in Roosevelt speech are', len(num_chars))
```

```python
num_sents = nltk.corpus.inaugural.sents('1941-Roosevelt.txt')
print('Number of sentences in Roosevelt speech are', len(num_sents))
```

###### Removing stop words from Roosevelt.txt

```python
all_words = nltk.corpus.inaugural.words('1941-Roosevelt.txt')
```

```python
# Defining a variable 'stopwords' which contains the list of punctuations from the string library and the english stopwords
# from nltk
stopwords = nltk.corpus.stopwords.words('english') +list(string.punctuation)

all_words = (x.lower() for x in all_words)
# Only keeping the words which are not the 'stopwords'
all_words_clean = [word for word in all_words if word not in stopwords]


# Creating a frequency distribution of the lower case words which does not contain any stopwords
all_words_freq = nltk.FreqDist(all_words_clean)
```

```python
all_words_clean
```

```python
stopwords
```

```python
nltk.FreqDist(all_words_clean)
```

```python
nltk.FreqDist(all_words_clean).most_common(4)
```

```python
word_features = [item[0] for item in all_words_freq.most_common()]
word_features
```

```python
from wordcloud import WordCloud
```

```python
text = all_words_clean
text
```

```python
from wordcloud import WordCloud
wordcloud = WordCloud(width = 3000, height = 3000, 
                background_color ='black', 
                min_font_size = 10, random_state=100).generate(str(text)) 
  
# plot the WordCloud image                        
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off")
plt.xlabel('Word Cloud')
plt.tight_layout(pad = 0) 

print("Word Cloud for Roosevelt Speech")
plt.show()
```
